```python
# DO NOT RUN THIS CODE TWICE

# Preprocessing the General Data 

# Required Modules
import numpy      as np
import pandas     as pd
from matplotlib import pyplot as plt
import seaborn    as sns

# Setting Font
plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
plt.rcParams['figure.figsize'] = (20, 6) # 그래프 크기 설정
plt.rc('font', size=20)        # 기본 폰트 크기
plt.rc('axes', labelsize=20)   # x,y축 label 폰트 크기
plt.rc('xtick', labelsize=20)  # x축 눈금 폰트 크기 
plt.rc('ytick', labelsize=20)  # y축 눈금 폰트 크기
plt.rc('legend', fontsize=20)  # 범례 폰트 크기
plt.rc('figure', titlesize=50) # figure title 폰트 크기

# Contants
NaN = 'NaN' # Not a Number

CITY_CODE = {
    11 : '서울특별시',
    26 : '부산광역시',
    27 : '대구광역시',
    28 : '인천광역시',
    29 : '광주광역시',
    30 : '대전광역시',
    31 : '울산광역시',
    36 : '세종특별자치시',
    41 : '경기도',
    42 : '강원도',
    43 : '충청북도',
    44 : '충청남도',
    45 : '전라북도',
    46 : '전라남도',
    47 : '경상북도',
    48 : '경상남도',
    49 : '제주특별자치도'
}

SEX_CODE = {
    1 : '남자',
    2 : '여자'
}

AGE_CODE = {
    1 : '소아',
    2 : '소아',
    3 : '10대',
    4 : '10대',
    5 : '20대',
    6 : '20대',
    7 : '30대',
    8 : '30대',
    9 : '40대',
    10 : '40대',
    11 : '50대',
    12 : '50대',
    13 : '60대',
    14 : '60대',
    15 : '70대',
    16 : '70대',
    17 : '80대 이상',
    18 : '80대 이상'
}

def get_hdl_c(hdl):
    if hdl < 40:
        return '비정상(저)'
    else:
        return '정상'
    
def get_ldl_c(ldl):
    if ldl < 130:
        return '정상'
    elif ldl < 160:
        return '비정상(약간높음)'
    elif ldl < 190:
        return '비정상(높음)'
    elif ldl >= 190:
        return '비정상(아주높음)'
    
def get_total_cholesterol_c(total_cholesterol):
    if total_cholesterol < 200:
        return '정상'
    elif total_cholesterol < 240:
        return '비정상(약간높음)'
    else:
        return '비정상(아주높음)'

def get_triglyceride_c(triglyceride):
    if triglyceride < 200:
        return '정상'
    elif triglyceride < 500:
        return '비정상(높음)'
    else:
        return '비정상(아주높음)'

def get_bmi_c(bmi):
    if bmi < 20:
        return '저체중'
    if bmi < 25:
        return '정상'
    if bmi < 30:
        return '과체중'
    if bmi >= 30:
        return '비만'

def get_alcohol_c(alcohol):
    if alcohol == 1:
        return '마신다'
    else:
        return '안마신다'
    
def get_smoke_c(smoke):
    if smoke == 1:
        return '비흡연'
    elif smoke == 2:
        return '금연'
    else:
        return '흡연'

def get_ear_c(ear):
    if ear == 1:
        return '정상'
    elif ear == 2:
        return '난청'
    else:
        return NaN

def get_ast_c(ast):
    if ast <= 32.0:
        return '정상'
    elif ast >= 51.0:
        return '질환의심자'
    else:
        return '위험'

def get_alt_c(alt):
    if alt <= 32.0:
        return '정상'
    elif alt >= 46.0:
        return '질환의심자'
    else:
        return '위험'

def get_proteinuria_c(proteinuria):
    if proteinuria == 1.0:
        return '음성'
    elif 2.0 <= proteinuria and proteinuria <= 6.0:
        return '양성'
    else:
        return NaN

def get_cavity_c(cavity):
    if cavity == 0.0:
        return '없음'
    elif cavity == 1.0:
        return '있음'
    else:
        return NaN

def get_scale_c(scale):
    if scale == 0.0:
        return '없음'
    elif scale == 1.0:
        return '있음'
    else:
        return NaN

def get_glucose_c(glucose):
    if glucose < 70:
        return '비정상(저)'
    elif glucose < 100:
        return '정상'
    elif glucose >= 100:
        return '비정상(고)'
    else:
        return NaN

def get_sbp_c(sbp):
    if sbp <= 100:
        return '저혈압'
    elif sbp >= 145:
        return '2차 진단 필요'
    else:
        return '정상'

def get_dbp_c(dbp):
    if dbp <= 60:
        return '저혈압'
    elif dbp >= 95:
        return '2차 진단 필요'
    else:
        return '정상'

def get_creatinine_c(row):
    if row['성별'] == '남자' and 0.61 <= row['혈청크레아티닌'] and row['혈청크레아티닌'] <= 1.04:
        return '정상'
    elif row['성별'] == '여자' and 0.47 <= row['혈청크레아티닌'] and row['혈청크레아티닌'] <= 0.79:
        return '정상'
    elif pd.isna(row['혈청크레아티닌']):
        return NaN
    else:
        return '비정상'

def get_hemoglobin_c(row):
    if row['성별'] == '남자' and row['혈색소'] < 14:
        return '비정상(저)'
    elif row['성별'] == '남자' and row['혈색소'] <=18:
        return '정상'
    elif row['성별'] == '여자' and row['혈색소'] < 12:
        return '비정상(저)'
    elif row['성별'] == '여자' and row['혈색소'] <=16:
        return '정상'
    else:
        return '비정상(고)'

def get_gamma_gtp_c(row):
    if row['성별'] == '남자' and 11 <= row['감마 지티피'] and row['감마 지티피'] <= 63:
        return '정상'
    elif row['성별'] == '여자' and 8 <= row['감마 지티피'] and row['감마 지티피'] <= 35:
        return '정상'
    else:
        return '비정상'

def get_blood_pressure_c(row):
    if row['수축기 혈압_c'] == '저혈압' or row['이완기 혈압_c'] == '저혈압':
        return '저혈압'
    elif row['수축기 혈압_c'] == '정상' and row['이완기 혈압_c'] == '정상':
        return '정상'
    else:
        return NaN

def process_nan(df):
    """
    결측치 및 이상치를 기준에 맞추어 처리한다.

    [Parameters]
    df (pandas.core.frame.DataFrame) : 결측치 및 이상치를 처리할 Pandas Dataframe

    [Returns]
    df (pandas.core.frame.DataFrame) : 결측치 및 이상치가 처리된 Pandas Dataframe
    """

    df['HDL 콜레스테롤'].fillna(df['HDL 콜레스테롤'].median(), inplace=True)
    df['LDL 콜레스테롤'].fillna(df['LDL 콜레스테롤'].median(), inplace=True)
    df['트리글리세라이드'].fillna(df['트리글리세라이드'].median(), inplace=True)
    df['청력(좌)'] = df['청력(좌)'].fillna(3.0).astype('int')
    df['청력(우)'] = df['청력(우)'].fillna(3.0).astype('int')
    df['시력(좌)'] = df['시력(좌)'].fillna(df['시력(좌)'].mode()[0]).astype('float')
    df['시력(우)'] = df['시력(우)'].fillna(df['시력(우)'].mode()[0]).astype('float')
    df.loc[df['시력(좌)']==9.9,'시력(좌)'] = 0.0
    df.loc[df['시력(우)']==9.9,'시력(우)'] = 0.0
    df['(혈청지오티)AST'].fillna(df['(혈청지오티)AST'].median(), inplace=True)
    df['(혈청지오티)ALT'].fillna(df['(혈청지오티)ALT'].median(), inplace=True)
    df['요단백'].fillna(0, inplace=True)
    df['치아우식증유무'].fillna('검진필요', inplace=True)
    df['치석'].fillna('검진필요', inplace=True)

    cond_m = df['성별'] == '남자'
    cond_f = df['성별'] == '여자'
    mean_m = df.groupby('성별')['혈색소'].mean()[0]
    mean_f = df.groupby('성별')['혈색소'].mean()[1]

    df.loc[cond_m, '혈색소'] = df.loc[cond_m, '혈색소'].fillna(mean_m)
    df.loc[cond_f, '혈색소'] = df.loc[cond_f, '혈색소'].fillna(mean_f)

    df.loc[df['감마 지티피'].isna(),'감마 지티피_c']= NaN
    df.loc[df['감마 지티피'].isna(),'감마 지티피']= 0.0
    df['수축기 혈압'] = df['수축기 혈압'].fillna(999)
    df['이완기 혈압'] = df['이완기 혈압'].fillna(999)
    
    return df

def categorize(df):
    """
    수치 데이터를 범주화하여 새로운 칼럼에 저장한다.

    [Parameters]
    df (pandas.core.frame.DataFrame) : 범주화되지 않은 Pandas Dataframe

    [Returns]
    df (pandas.core.frame.DataFrame) : 범주화된 Pandas Dataframe
    """

    # Apply Function with Single Arguments
    df['HDL 콜레스테롤_c'] = df['HDL 콜레스테롤'].apply(get_hdl_c)
    df['LDL 콜레스테롤_c'] = df['LDL 콜레스테롤'].apply(get_ldl_c)
    df['총 콜레스테롤_c'] = (df['총 콜레스테롤']).apply(get_total_cholesterol_c)
    df['트리글리세라이드_c'] = df['트리글리세라이드'].apply(get_triglyceride_c)
    df["BMI_c"] = df["BMI"].apply(get_bmi_c)
    df['음주여부_c'] = df['음주여부'].apply(get_alcohol_c)
    df['흡연상태_c'] = df['흡연상태'].apply(get_smoke_c)
    df['청력(좌)_c'] = df['청력(좌)'].apply(get_ear_c)
    df['청력(우)_c'] = df['청력(우)'].apply(get_ear_c)
    df['(혈청지오티)AST_c'] = df['(혈청지오티)AST'].apply(get_ast_c)
    df['(혈청지오티)ALT_c'] = df['(혈청지오티)ALT'].apply(get_alt_c)
    df['요단백_c'] = df['요단백'].apply(get_proteinuria_c)
    df['치아우식증유무_c'] = df['치아우식증유무'].apply(get_cavity_c)
    df['치석_c'] = df['치석'].apply(get_scale_c)
    df['식전혈당_c'] = df['식전혈당(공복혈당)'].apply(get_glucose_c)
    df['수축기 혈압_c'] = df['수축기 혈압'].apply(get_sbp_c)
    df['이완기 혈압_c'] = df['이완기 혈압'].apply(get_dbp_c)

    # Apply Function with Multiple Arguments
    df['혈청크레아티닌_c'] = df.apply(func=get_creatinine_c, axis=1)
    df['혈색소_c'] = df.apply(func=get_hemoglobin_c, axis=1)
    df['감마 지티피_c'] = df.apply(func=get_gamma_gtp_c, axis=1)
    df['혈압 상태_c'] = df.apply(func=get_blood_pressure_c, axis=1)
    
    return df

def preprocess(df):
    """
    데이터를 전처리한다.

    [Parameters]
    df (pandas.core.frame.DataFrame) : 전처리되지 않은 Pandas Dataframe

    [Returns]
    df (pandas.core.frame.DataFrame) : 전처리된 Pandas Dataframe
    """

    code_table = [CITY_CODE, SEX_CODE, AGE_CODE]
    old_columns = ['시도코드', '성별코드', '연령대 코드(5세단위)']
    new_columns = ['시도', '성별', '연령대']
    useless_columns = ['가입자 일련번호', '기준년도', '데이터 공개일자', '구강검진 수검여부']

    # Insert Preprocessed Columns
    for (new_column, old_column, code) in zip(new_columns, old_columns, code_table):
        df[new_column] = df[old_column].map(code)
        df = df.drop(labels=old_column, axis=1)

    # Column Relocation
    new_cols = df.columns[-1*len(new_columns):].to_list()
    exist_cols = df.columns[:-1*len(new_columns)].to_list()
    df = df[new_cols+exist_cols]

    # Reset Index
    df = df.reset_index(drop=True)

    # Calculate New Column (Before Processing NaN)
    df["BMI"] = round(df['체중(5Kg 단위)'] / (df['신장(5Cm단위)']/100)**2,1)
    df['치아건강'] = '비정상'

    cond_scale = (df['치석'] == '없음')
    cond_cavity = (df['치아우식증유무'] == '없음')
    df.loc[cond_scale & cond_cavity, '치아건강'] = '정상'
    
    cond_absence = df['구강검진 수검여부'] == 0
    df.loc[cond_absence, '치아건강'] = NaN

    # Drop Useless Columns
    for useless_column in useless_columns:
        df = df.drop(labels=useless_column, axis=1)

    # Categorization
    df = categorize(df)

    # Handling Missing Data
    df = process_nan(df)

    # Calculate New Column (After Processing NaN)
    df['시력(평균)'] = (df['시력(좌)'] + df['시력(우)'])/2
    df.loc[df['시력(좌)'] == 0,'시력(평균)'] = df.loc[df['시력(좌)'] == 0,'시력(우)']
    df.loc[df['시력(우)'] == 0,'시력(평균)'] = df.loc[df['시력(우)'] == 0,'시력(좌)']
    df['시력(차이)'] = abs(df['시력(좌)'] - df['시력(우)'])

    return df

df = preprocess(검진정보)
pd.set_option('display.max_columns', None)
df
```

```python
df['건강점수'] = 100
score1 = df['(혈청지오티)AST_c'] == '정상'
score2 = df['(혈청지오티)ALT_c'] == '정상'
score3 = df['감마 지티피_c'] == '정상'

score4 = df['총 콜레스테롤_c'] == '정상'
score5 = df['트리글리세라이드_c'] == '정상'
score6 = df['HDL 콜레스테롤_c'] == '정상'
score7 = df['LDL 콜레스테롤_c'] == '정상'

score8 = df['BMI_c'] == '정상'

score9 = df['식전혈당_c'] == '정상'

score10 = df['혈압 상태_c'] =='정상'

# 간 기능 건강점수
df.loc[~score1, '건강점수'] = df['건강점수'] - 6
df.loc[~score2, '건강점수'] = df['건강점수'] - 7
df.loc[~score3, '건강점수'] = df['건강점수'] - 7

# 콜레스테롤 수치 건강점수
df.loc[~score4, '건강점수'] = df['건강점수'] - 5
df.loc[~score5, '건강점수'] = df['건강점수'] - 5
df.loc[~score6, '건강점수'] = df['건강점수'] - 5
df.loc[~score7, '건강점수'] = df['건강점수'] - 5

# 비만 건강점수
df.loc[~score8, '건강점수'] = df['건강점수'] - 20

# 혈당 건강점수
df.loc[~score9, '건강점수'] = df['건강점수'] - 20

# 혈압 건강점수
df.loc[~score10, '건강점수'] = df['건강점수'] -20

#건강 점수별 분류
def scoring(x):
    if x < 40:
        return '즉각 치료 요망'
    elif x <= 60:
        return '심각'
    elif x <=80:
        return '주의 요망'
    else:
        return '건강'
    
df['건강상태'] = df['건강점수'].apply(scoring)
```
