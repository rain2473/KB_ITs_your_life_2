### concat

- 위의 결과를 반복문을 사용하여 각 파일을 데이터프레임으로 읽어서 data 리스트에 저장하세요.
    - encoding = 'cp949'
    - data 결과 : `[데이터프레임1, 데이터프레임2, 데이터프레임3, ...]`

```python
data = []

for file_name in os.listdir(path) :
    file_name = path + file_name
    # file_name 파일을 데이터프레임으로 읽어 data 리스트에 추가하세요.
    data.append(pd.read_csv(file_name,encoding='cp949'))

# 첫 번째 데이터 프레임을 조회하세요.
data[0]
```

```python
# data 리스트의 길이를 구하세요.
len(data)
```

```python
# 12개의 데이터프레임을 행 방향으로 연결하세요.
df = pd.concat(data,axis = 0)
df
```

### 시계열 데이터 다루기

```python
# 각 열 별 데이터 타입을 조회하세요.
df.dtypes
```

```python
# '기간' 열에서 문자열 인덱싱, 슬라이싱을 이용하여
# 연도 정보를 'year' 열에, 월 정보를 'month' 열에, 주 정보를 'week' 열에 추가하세요.
df['year'] = df['기간'].str[0:4]
df['month'] = df['기간'].str[6:8]
df['week'] = df['기간'].str[10]
df.head()
```

```python
# '기간' 열을 삭제하세요.
df = df.drop('기간',axis=1)
df.head()
```

```python
# 'week' 열의 데이터 타입을 'int32'로 변경하세요.
df['week'] = df['week'].astype('int32')
df.dtypes
```

```python
# 'week' 열에서 1주차는 1일, 2주차는 8일, 3주차는 15일, 4주차는 22일, 5주차는 29일 정보를 'day' 열에 추가하세요.
# apply 사용 (replace 사용 x)
def day(x):
    tmp = [1,8,15,22,29]
    return tmp[x-1]
df['day'] = df['week'].apply(day)
df.head()
```

```python
# 'day' 열의 데이터 타입을 'string'으로 변경하세요.
df['day'] = df['day'].astype('string')
df.dtypes
```

```python
# 코드를 실행하세요.
# 2자리에 맞춰 '0'으로 채우는 함수 ex) '1' -> '01'
df['day'] = df['day'].str.zfill(2)
```

```python
# 'year', 'month', 'day' 열을 조합하여 '2021-01-01' 과 같은 포맷을 '기간' 열에 추가하세요.
df['기간'] = df['year']+'-'+df['month']+'-'+df['day']
df.head()
```

```python
# '기간' 열의 데이터 타입을 'datetime64'로 변경하세요.
df['기간'] = df['기간'].astype('datetime64')
df.dtypes
```

```python
# '기간' 열에서 요일 정보를 추출하여 '요일' 열에 추가하세요.
# 요일 정보 : 0은 월요일
df['요일'] = df['기간'].dt.dayofweek
df.head()
```

```python
# '기간' 별 '휘발유' 가격의 평균을 df1 에 저장하세요.
df1 = df.pivot_table(index=['기간'],values=['휘발유'])
df1
```

```python
# df1의 인덱스를 초기화하세요.
# drop = False
df1 = df1.reset_index()
df1
```

- (참고) 시각화

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.figure(figsize = (12,6))

sns.lineplot(data = df1, x = '기간', y = '휘발유', color = 'orange')


plt.axhline(df1['휘발유'].mean(), color = 'gray', ls = '--', label = '평균가격')
plt.text(df1.loc[0, '기간'], df1['휘발유'].mean(), f"{df1['휘발유'].mean():.0f} 원",
         fontsize = 16, color = 'gray', ha = 'left', va = 'bottom')

plt.axhline(df1['휘발유'].max(), color = 'red', ls = '--', label = '최대값')
plt.text(df1.loc[0, '기간'], df1['휘발유'].max(), f"{df1['휘발유'].max():.0f} 원",
         fontsize = 16, color = 'red', ha = 'left', va = 'top')

plt.axhline(df1['휘발유'].min(), color = 'blue', ls = '--', label = '최소값')
plt.text(df1.loc[0, '기간'], df1['휘발유'].min(), f"{df1['휘발유'].min():.0f} 원",
         fontsize = 16, color = 'blue', ha = 'left', va = 'bottom')

plt.legend(loc = 'lower right')
plt.show()
```

### merge

```python
# 코드를 실행하세요.
# 중간고사 결과를 생성하세요.
midterm = pd.DataFrame({
    'name' : ['Kim', 'Park', 'Lee', 'Choi'],
    'midterm' : [100, 50, 50, 80]
})
midterm
```

```python
# 코드를 실행하세요.
# 기말고사 결과를 생성하세요.
# 기말고사에는 'Lee'가 전학을 가고, 'New'가 전학을 왔습니다.
final = pd.DataFrame({
    'name' : ['Choi', 'Park', 'New', 'Kim'],
    'final' : [90, 40, 95, 70]
})
final
```

```python
# 중간고사, 기말고사 정보를 이름을 기준으로 병합하세요.
# 두 시험을 모두 본 학생들의 정보만 병합하세요.
pd.merge(midterm,final,how='inner')
```

```python
# 중간고사, 기말고사 정보를 이름을 기준으로 병합하세요.
# 모든 학생들의 정보를 남기세요.
pd.merge(midterm,final,how='outer')
```

- 동명이인이 전학 온 경우

```python
# 코드를 실행하세요.
# 중간고사 결과를 생성하세요.
midterm = pd.DataFrame({
    'name' : ['Kim', 'Park', 'Lee', 'Choi'],
    'midterm' : [100, 50, 50, 80]
})
midterm
```

```python
# 코드를 실행하세요.
# 기말고사 결과를 생성하세요.
# 기말고사에는 'Lee'가 전학을 가고, 'Park'이 전학을 왔습니다.
final = pd.DataFrame({
    'name' : ['Choi', 'Park', 'Park', 'Kim'],
    'final' : [90, 40, 95, 70]
})
final
```

```python
# 중간고사, 기말고사 정보를 이름을 기준으로 병합하세요.
# 두 시험을 모두 본 학생들의 정보만 병합하세요.
pd.merge(midterm,final,how='inner')
```

```python
# 코드를 실행하세요.
# 새로 전학 온 'Park' 의 중간고사 점수가 기존의 'Park'의 점수로 기입되었습니다.
# 이름이 같은 학생을 구분하기 위해 'birth' 정보를 추가하세요.
midterm = pd.DataFrame({
    'name' : ['Kim', 'Park', 'Lee', 'Choi'],
    'birth' : ['0101', '1225', '0228', '1031'],
    'midterm' : [100, 50, 50, 80]
})
final = pd.DataFrame({
    'name' : ['Choi', 'Park', 'Park', 'Kim'],
    'birth' : ['1031', '1225', '0505', '0101'],
    'final' : [90, 40, 95, 70]
})

display(midterm)
display(final)
```

```python
# 중간고사, 기말고사 정보를 이름과 생일을 기준으로 병합하세요.
# 두 시험을 모두 본 학생들의 정보만 병합하세요.
pd.merge(midterm,final,how='inner')
```

```python
# 중간고사, 기말고사 정보를 이름과 생일을 기준으로 병합하세요.
# 모든 학생들의 정보를 남기세요.
pd.merge(midterm,final,how='outer')
```
