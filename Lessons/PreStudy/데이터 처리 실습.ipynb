```python
import numpy as np
import pandas as pd
data = pd.read_csv('data/president_heights.csv')
heights = np.array(data['height(cm)'])
names = np.array(data['name'])
print(heights)
```

```python
print(round(heights.mean(),2))
print(round(heights.std(),2))
print(round(heights.var(),2))
print(heights.max())
print(names[heights.argmax()])
print(heights.min())
print(names[heights.argmin()])
print(np.percentile(heights,25))
print(np.median(heights))
print(np.percentile(heights,75))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.hist(heights)
plt.title('height Distribution of US Presidents')
plt.xlabel('height(cm)')
plt.ylabel('number')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
plt.hist(heights)
plt.title('height Distribution of US Presidents')
plt.xlabel('height(cm)')
plt.ylabel('number')
```
