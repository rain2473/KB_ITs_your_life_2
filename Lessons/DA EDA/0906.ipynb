## 인덱서 
##### df.loc[행인덱스, 열인덱스]
##### loc -> index(행), column name(열) (통칭)index (대표값)
##### 행인덱스 -> 인덱싱, 팬시인덱싱, 슬라이싱, 조건
##### 열인덱스 -> 인덱싱, 팬시인덱싱, 슬라이싱, 조건
##### 값 변경시 복제가 아닌 원본 수정
##### iloc -> 순번만 사용
##### 행인덱스 -> 인덱싱, 팬시인덱싱, 슬라이싱, 조건
##### 열인덱스 -> 인덱싱, 팬시인덱싱, 슬라이싱, 조건

## sort

```python
ti.sort_values('age').head()
```

## iloc
loc과 유사하나 인덱스 숫자값만 사용 가능
인덱싱, 슬라이싱 모두 가능함

```python
print(ti.sort_values('age').iloc[0])     # 조건내에서 첫번째 항목, 순번임          // 정렬에 따라 순번이 바뀜
print(ti.sort_values('age').loc[0])      # 원래 DF에서 첫번째 항목, 인덱스 번호임  // 순번이 유지된다.
```

```python
# at 하나의 인덱스만 가져옴
%timeit ti.loc[0,'fare']
%timeit ti.at[0,'fare']
```

## groupby

```python
ti.groupby('sex').mean()
```

```python
ti.groupby('sex')['survived'].mean()
ti.groupby('sex')['age'].mean()
```

```python
# 성별 좌석등급별 생존율
display(ti.groupby(['sex','pclass'])['survived'].mean())
# 좌석등급별 성별 생존율
display(ti.groupby(['pclass','sex'])['survived'].mean())
```

```python
성별등급별S = ti.groupby(['sex','pclass'])['survived'].mean()
성별등급별S
```

```python
성별등급별DF = ti.groupby(['sex','pclass'])[['survived']].mean()
성별등급별DF
```

```python
# 성별등급별 1등석 여성승객의 생존율 출력
성별등급별S.loc[('female',1)]
```

```python
# 성별등급별 1등석 여성승객의 생존율 출력
성별등급별DF.loc[('female',1),:]
```

```python
#reset_index()
성별등급별 = ti.groupby(['sex','pclass'])['survived']
성별등급별.mean().reset_index()
```

```python
ti.groupby(['sex','pclass'])[['survived','age']].agg(['mean','std']) # columns은 리스트로 줘야함 ㅈ=> 여러개면 리스트라 생각하자
```

## pivot_table

```python
ti.pivot_table(index='who',values='survived')
```

```python
ti.pivot_table(columns='who',values='survived')
```

```python
ti.pivot_table(
    index=['who','pclass'],
    values='survived'
)
```

```python
ti.pivot_table(
    index=['who'],
    columns=['pclass'],
    values='survived'
)
```

```python
ti.pivot_table(
    index=['who'],
    columns=['pclass'],
    values='survived',
    aggfunc=['sum','mean']
)
```
