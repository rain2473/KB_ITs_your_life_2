```python
# apt_price['시군구']를 시,구,동으로 분할
tmp = apt_price['시군구'].str.split(' ')
apt_price['구'] = tmp.str[1]
apt_price['동'] = tmp.str[2]
```

```python
# 단위 변환 
apt_price['평'] = round(apt_price['전용면적(㎡)'] / 3.03, 1)         #전용면적을 평 단위로 변환
apt_price['거래금액(억원)'] = apt_price['거래금액(만원)'] /10000    #거래금액을 억 단위로 변환
```

```python
# 평형 구간 나누기
tmp = [0,10,20,30,40,50,1000000]
tmp1 = '10평미만,10평대,20평대,30평대,40평대,50평이상'.split(',')
apt_price['평형'] = pd.cut(apt_price['평'], bins = tmp, right = False, labels=tmp1)
```

```python
# 계약년월 분할하기
apt_price['계약년'] = apt_price['계약년월'] // 100
apt_price['계약월'] = apt_price['계약년월'] % 100
```

```python
# 계약 요일 추출 - 1
tmp = apt_price['계약년월'].astype(str)+apt_price['계약일'].astype(str)
apt_price['계약요일'] = pd.to_datetime(tmp, format='%Y%m%d').dt.dayofweek
apt_price['계약요일']
```

```python
# 계약 요일 추출 - 2
def dow(i):
    week_list = '월,화,수,목,금,토,일'.split(',')
    return week_list[i]+'요일'
apt_price['계약요일'] = apt_price['계약요일'].apply(dow)
apt_price['계약요일']
```

```python
apt_전세 = pd.read_csv(
    '아파트전월세.csv',
    encoding = 'cp949',             # 한글 문자셋
    skiprows=15,                    # 설명부분 삭제
    thousands=','                   # 천단위 구분자
)
전세 = apt_전세[apt_전세['전월세구분']=='전세'].copy()
월세 = apt_전세[apt_전세['전월세구분']=='월세'].copy()

def 전환률적용(x):
    전환률 = 4.75 /100
    return (x*12/전환률)/10000

월세['보증금(억원)'] = 월세['월세(만원)'].apply(전환률적용) + 월세['보증금(만원)']/10000
월세['보증금(억원)']
전세['보증금(억원)'] = 전세['보증금(만원)']/10000

apt_전세=pd.concat([월세,전세]).sort_index()
apt_전세['보증금(억원)'] = round(apt_전세['보증금(억원)'],2)

# apt_전세['시군구']를 시,구,동으로 분할
tmp = apt_전세['시군구'].str.split(' ')
apt_전세['구'] = tmp.str[1]
apt_전세['동'] = tmp.str[2]

# 단위 변환 
apt_전세['평'] = round(apt_전세['전용면적(㎡)'] / 3.03, 1)         #전용면적을 평 단위로 변환

# 평형 구간 나누기
tmp = [0,10,20,30,40,50,1000000]
tmp1 = '10평미만,10평대,20평대,30평대,40평대,50평이상'.split(',')
apt_전세['평형'] = pd.cut(apt_전세['평'], bins = tmp, right = False, labels=tmp1)

# 계약년월 분할하기
apt_전세['계약년'] = apt_전세['계약년월'] // 100
apt_전세['계약월'] = apt_전세['계약년월'] % 100
```
