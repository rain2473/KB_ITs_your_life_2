```python
# 거래건수가 가장 많은 구의 이름과 건수
구_list = list(apt_price['구'].unique())
sep = []
cnt = 0

for i in range(len(구_list)):
    sep.append(apt_price[apt_price['구'] == str(구_list[i])].shape[0])
for j in range(len(sep)):
    tmp1 = 0
    if sep[j] > tmp1:
        tmp1 = sep[j]
        cnt += 1
cnt
```

```python
#거래건수가 제일많은 구 이름과 건수를 출력하세요.
max_name = ""
max_count = 0
# 사용된 구이름 목록 추출(구칼럼.unique())
구이름 = apt_price["구"].unique()
# 구이름을 반복(for)
for 구 in 구이름:
#     print(구,apt_price[apt_price['구'] == 구].shape)
    거래건수 = apt_price[apt_price['구'] == 구].shape[0]
    if 거래건수 > max_count:
        max_count = 거래건수
        max_name = 구
print("거래건수가 가장많은 구",max_name, max_count,'건')
```

```python
# 평방미터를 평으로 변환
apt_price['평'] = apt_price['전용면적(㎡)'] / 3.03
apt_price['평']
```

```python
# 평형 기준 구간으로 나누기
apt_price['평형'] = pd.cut(
    apt_price['평'],
    bins = [0,10,20,30,40,50, 10000],
    right = False,
    labels = ['10평 미만', '10평대', '20평대', '30평대', '40평대', '50평 이상']
)
```

## groupby()
#### 매우중요 - 그룹별 
pivottable과 유사

```python
apt_price.groupby('구').mean()
```

```python
apt_price.groupby('구')['거래금액(만원)'].describe()
```

```python
apt_price['거래금액(만원)'].describe()
```

```python
# 거래금액(만원)을 기준으로 1억미만, 5억미만, 10억 미만, 20억 미만, 50억 미만, 50억 이상으로 구간분할
apt_price['가격대'] = pd.cut(
    apt_price['거래금액(만원)'],
    bins = [0,10000,50000,100000,200000,500000,100000000],
    right = False,
    labels = '1억미만, 5억미만, 10억 미만, 20억 미만, 50억 미만, 50억 이상'.split(',')
)
apt_price
```

```python
# 가격대별 거래건수
apt_price.groupby('가격대')['거래금액(만원)'].count()
```

```python
# 강남3구의 평형별 평균 거래가
# 이외 지역의 평형별 평균 거래가
cond = apt_price['구'].isin(['강남구','서초구','송파구'])
display(apt_price[cond].groupby('평형')['거래금액(만원)'].mean())  #3구
display(apt_price[~cond].groupby('평형')['거래금액(만원)'].mean()) #이외
display(apt_price.groupby('평형')['거래금액(만원)'].mean())        #전체
```
