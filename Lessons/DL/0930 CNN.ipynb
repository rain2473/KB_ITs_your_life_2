```python
#  
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input,decode_predictions
import numpy as np
import glob
import matplotlib.pyplot as plt
import math

#   
test_dir_path = './img/others/custom/*.*'

# test  
test_paths = glob.glob(test_dir_path)
print(test_paths)
```

```python
#    
plt.figure(figsize=(10,10))

#       
for i, path in enumerate(test_paths):
    #          (path =   )
    img = tf.keras.preprocessing.image.load_img(path, target_size=(IMG_SIZE, IMG_SIZE))
    #   
    img_arr = tf.keras.preprocessing.image.img_to_array(img)#, dtype=np.uint8)
    #   3 => 4 (224,224,3 ==> 1,224,224,3  0    )
    img_arr = tf.expand_dims(img_arr, 0)

    #   vgg16 predict  ==> vgg16  1      
    pred = vgg16.predict(preprocess_input(img_arr))
    # pred     5   (   (, , )) ==> top    
    decoded = decode_predictions(pred)[0]
    #  4 subplot 
    plt.subplot(math.ceil(len(test_paths)/4), 4, i+1)
    #  
    plt.imshow(img)
    #   decoded(  )  
    plt.title(f'{decoded[0][1]}({decoded[0][2]*100:.2f}%)')
    #   
    plt.axis('off')
plt.show()
```

```python
#  
import zipfile
import os

# 
dir_path = 'dogs-vs-cats'
```

```python
#    
if not os.path.exists(dir_path):
    os.makedirs(dir_path)

#  
with zipfile.ZipFile("dogs-vs-cats.zip" , "r") as zip_ref:
    zip_ref.extractall(dir_path)
    
#    
with zipfile.ZipFile(dir_path+"/train.zip" , "r") as zip_ref:
    zip_ref.extractall(dir_path)
```

```python
#   
sub_dirs = ['train', 'test']
label_dirs = ['cat', 'dog']

#  
for sub_dir in sub_dirs:
    for label_dir in label_dirs:
        os.makedirs(os.path.join(dir_path, sub_dir, label_dir), exist_ok=True)
```

```python
import glob
import random
import shutil

files = glob.glob(dir_path + '/train/*.jpg')
split_ratio = 0.25

#   
for file in files:
    #     
    filename = os.path.basename(file)
    print('file copying...', file, end='\r')
    #   train
    dst_dir = 'train'
    #  0.25  test, cat dog   
    if random.random() < split_ratio :
        dst_dir = 'test'
    if filename.startswith('cat'):
        dst = os.path.join(dir_path, dst_dir, 'cat', filename)
    else :
        dst = os.path.join(dir_path, dst_dir, 'dog', filename)
    shutil.copyfile(file, dst)
print('copy done.')
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
#   => 
train_datagen = ImageDataGenerator(
      rescale=1./255,
      rotation_range=20,
      width_shift_range=0.1,
      height_shift_range=0.1,
      shear_range=0.1,
      zoom_range=0.1,
      horizontal_flip=True,
      vertical_flip= True,
      fill_mode='nearest')
test_datagen = ImageDataGenerator(rescale=1./255) # ,  
train_generator = train_datagen.flow_from_directory(
        os.path.join(dir_path, 'train'), #  
        target_size=(150, 150),#    150  150 
        batch_size=20,
        class_mode='binary')
validation_generator = test_datagen.flow_from_directory(
        os.path.join(dir_path, 'test'),
        target_size=(150, 150),
        batch_size=20,
        class_mode='binary')
```

```python
image_batch, label_batch = next(train_generator)

plt.figure(figsize=(10,10))

# image_batch 0~19 
for n in range(20):
    plt.subplot(5,5,n+1)
    plt.imshow(image_batch[n])
    plt.title(label_batch[n])
    plt.axis('off')
```

```python
from tensorflow.keras import optimizers
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras.applications import VGG16

IMG_SIZE = 150
vgg16 = VGG16(weights='imagenet',
             include_top = False,
             input_shape = (IMG_SIZE, IMG_SIZE, 3))

model = models.Sequential()
model.add(vgg16)
model.add(layers.Flatten())
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(lr=2e-5), # argument is deprecated. use learning_rate=2e-5
              metrics=['acc'])

history = model.fit_generator(
      train_generator,
      steps_per_epoch=100,
      epochs=30,
      validation_data=validation_generator,
      validation_steps=50,
      verbose=2)
```
