```python
import numpy as np
import tensorflow as tf
import pandas as pd

n_inpu =28*28
n_L1 = 300
n_L2 = 100
n_output = 10

(X_tr, y_tr),(X_te,y_te) = tf.keras.datasets.mnist.load_data()
```

```python
X_va, X_tr = X_tr[:5000], X_tr[5000:] # 검증용 5천, 학습용 5만5천
y_va, y_tr = y_tr[:5000], y_tr[5000:] # 검증용 5천, 학습용 5만5천
```

```python
X_va.shape, y_va.shape
```

```python
# 모델 선언
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, 3, activation='relu', input_shape=(28,28,1)), # 이동간격 stride  필터 32회 커널 사이즈 3*3
    tf.keras.layers.MaxPool2D(2, 2), # MAXPool 시행, 풀 사이즈 2 stride 2
    tf.keras.layers.Flatten(),      # 입력을 1차원으로 만듬
    tf.keras.layers.Dense(128, activation=tf.nn.relu ), 
    tf.keras.layers.Dropout(0.3),   # 30퍼 버리기 => 과적합 경계
    tf.keras.layers.Dense(n_output, activation=tf.nn.softmax)   # 결과 10개 미리 선언, 다항분류 => softmax
])
```

```python
# 모델 컴파일링
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# 모델 fit
model.fit(X_tr, y_tr, epochs=10, batch_size=100, validation_data=(X_va, y_va))  
# 6만개 100개씩 잘라서 10번 반복
# 분할 데이터 직접 줌. => 내부 훈련 데이터를 외부 데이터로 가능.
```

```python
# 모델 평가
test_loss, test_acc = model.evaluate(X_te, y_te, verbose=0)
```

```python
# 테스트 점수
print("Test Accuracy:{}".format(test_acc))
```

```python
# 모듈 임포팅
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input,decode_predictions
import numpy as np
import glob
import matplotlib.pyplot as plt
import math

# 이미지 경로 지정
test_dir_path = './img/others/custom/*.*'

# test 경로 지정
test_paths = glob.glob(test_dir_path)
print(test_paths)
```

```python
# 경로 안의 사진 출력
plt.figure(figsize=(10,10))

# 경로 안의 모든 파일에 대하여 반복문 실행
for i, path in enumerate(test_paths):
    # 경로에서 이미지를 가져와서 이미지 사이즈 조절 후 객체로 저장 (path = 이미지 파일 경로)
    img = tf.keras.preprocessing.image.load_img(path, target_size=(IMG_SIZE, IMG_SIZE))
    # 이미지를 배열로 변환
    img_arr = tf.keras.preprocessing.image.img_to_array(img)#, dtype=np.uint8)
    # 차원 확장 3 => 4차원 (224,224,3 ==> 1,224,224,3  0때문에 맨 앞자리에 차원 추가)
    img_arr = tf.expand_dims(img_arr, 0)

    # 앞에서 선언한 vgg16을 predict 함 ==> vgg16에 저장된 1천개 이미지 종류 각각에 대한 확률 반환
    pred = vgg16.predict(preprocess_input(img_arr))
    # pred 중 확률이 높은 상위 5개만 남김 그중 첫번째(가장 일치율 높은것 반환(명칭, 이름, 확률)) ==> top 주면 갯수 조절 가능
    decoded = decode_predictions(pred)[0]
    # 한줄에 4개씩 subplot 출력
    plt.subplot(math.ceil(len(test_paths)/4), 4, i+1)
    # 이미지 출력
    plt.imshow(img)
    # 제목은 위의 decoded(일치율 가장 높은것)의 이름과 확률
    plt.title(f'{decoded[0][1]}({decoded[0][2]*100:.2f}%)')
    # 축 표시 끄기
    plt.axis('off')
plt.show()
```

```python
# 모델 임포팅
import zipfile
import os

# 경로설정
dir_path = 'dogs-vs-cats'
```

```python
# 경로 없으면 경로 만들기
if not os.path.exists(dir_path):
    os.makedirs(dir_path)

# 압축파일 열기
with zipfile.ZipFile("dogs-vs-cats.zip" , "r") as zip_ref:
    zip_ref.extractall(dir_path)
    
# 압축파일 속 압축파일 열기
with zipfile.ZipFile(dir_path+"/train.zip" , "r") as zip_ref:
    zip_ref.extractall(dir_path)
```

```python
# 경로 리스트 만들기
sub_dirs = ['train', 'test']
label_dirs = ['cat', 'dog']

# 경로 생성
for sub_dir in sub_dirs:
    for label_dir in label_dirs:
        os.makedirs(os.path.join(dir_path, sub_dir, label_dir), exist_ok=True)
```

```python
import glob
import random
import shutil

files = glob.glob(dir_path + '/train/*.jpg')
split_ratio = 0.25

# 모든 파일에 대하여
for file in files:
    # 경로의 맨 끝 이름을 추출
    filename = os.path.basename(file)
    print('file copying...', file, end='\r')
    # 기본 경로 train
    dst_dir = 'train'
    # 난수가 0.25보다 작으면 test, cat dog 분류해서 경로에 저장
    if random.random() < split_ratio :
        dst_dir = 'test'
    if filename.startswith('cat'):
        dst = os.path.join(dir_path, dst_dir, 'cat', filename)
    else :
        dst = os.path.join(dir_path, dst_dir, 'dog', filename)
    shutil.copyfile(file, dst)
print('copy done.')
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
# 데이터를 변형시킴 => 왜곡
train_datagen = ImageDataGenerator(
      rescale=1./255,
      rotation_range=20,
      width_shift_range=0.1,
      height_shift_range=0.1,
      shear_range=0.1,
      zoom_range=0.1,
      horizontal_flip=True,
      vertical_flip= True,
      fill_mode='nearest')
test_datagen = ImageDataGenerator(rescale=1./255) #검증 데이타, 증식 없슴
train_generator = train_datagen.flow_from_directory(
        os.path.join(dir_path, 'train'), # 타깃 디렉터리
        target_size=(150, 150),# 모든 이미지의 크기를 150 × 150로 변경
        batch_size=20,
        class_mode='binary')
validation_generator = test_datagen.flow_from_directory(
        os.path.join(dir_path, 'test'),
        target_size=(150, 150),
        batch_size=20,
        class_mode='binary')
```

```python
image_batch, label_batch = next(train_generator)

plt.figure(figsize=(10,10))

# image_batch의 0~19 까지
for n in range(20):
    plt.subplot(5,5,n+1)
    plt.imshow(image_batch[n])
    plt.title(label_batch[n])
    plt.axis('off')
```

```python
from tensorflow.keras import optimizers
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras.applications import VGG16

IMG_SIZE = 150
vgg16 = VGG16(weights='imagenet',
             include_top = False,
             input_shape = (IMG_SIZE, IMG_SIZE, 3))

model = models.Sequential()
model.add(vgg16)
model.add(layers.Flatten())
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(lr=2e-5), # argument is deprecated. use learning_rate=2e-5
              metrics=['acc'])

history = model.fit_generator(
      train_generator,
      steps_per_epoch=100,
      epochs=30,
      validation_data=validation_generator,
      validation_steps=50,
      verbose=2)
```
