```python
# 데이터 정의
# 관리대상: 연락처
#속성 - 필) 이름(문자열), 전화번호(문자열)
#       선) 이메일(문자열), 설명(문자열)
class Contact:
    def __init__(self,name,tel,mail = None, note=None):
        self.name = name
        self.tel = tel
        self.mail = mail
        self.note = note
        
    def __str__(self):
        return f'{self.name}{self.tel},{self.mail},{self.note}'
        
#c1 = Contact('홍길동', '010-4235-9847')
#c2 = Contact('홍길은', '010-4284-9847','hong@naver.com','친구')
#c3 = Contact('홍길금', '010-4874-9847', 'honggold@naver.com')
#c4 = Contact('홍길철', '010-4574-9847', '형제')
#print(f"{c1}\n{c2}\n{c3}\n{c4}")
```

```python
#데이터 관리자 객체(manager)
class ContactManager:
    def __init__(self):
        self.contacts = []
    
    def add(self,c):
        self.contacts.append(c)
        return [f'{c.name} : {c.tel}(이)가 등록되었습니다!']
    
    def count(self):
        return len(self.contacts)
    
    def con_list(self,target = "name"):
        ans = []
        for c in self.contacts:
            if target == 'name':
                ans.append(c.name)
            elif target == 'tel':
                ans.append(c.tel)
        return ans
    
    def search(self, keyword, target = 'name'):#주소가 반환됨 -> 수정
            ans = []
            for c in self.contacts:
                if target == 'name':
                    if keyword in c.name:
                        ans.append(c)
                elif target == 'tel':
                    if keyword in c.tel:
                        ans.append(c)
            if ans == []:
                ans = ['검색어를 찾을 수 없습니다.']
            
            return ans()
        
    def delete(self, keyword, target = 'name'):
            ans = []
            for c in self.contacts:
                if target == 'name':
                    if keyword == c.name:
                        ans = [f'{c.name} : {c.tel} 삭제에 성공하였습니다.']
                        self.contacts.remove(c)
                elif target == 'tel':
                    if keyword == c.tel:
                        ans = [f'{c.name} : {c.tel} 삭제에 성공하였습니다.']
                        self.contacts.remove(c)
            if ans == []:
                ans = ['검색어를 찾을 수 없습니다.']
                
            return ans
```
