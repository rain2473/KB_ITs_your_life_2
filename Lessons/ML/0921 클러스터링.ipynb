# 클러스터링

비지도 학습 : 주어진 데이터에 대한 결과가 없는 데이터를 이용해 주어진 데이터에 내재된 패턴, 특성, 구조를 찾아서 학습  

## 군집분석

특성이 유사한 것끼리 묶음  
분류 분석과 동일하게 데이터는 하나의 군집에 할당  
같은 군집 내 데이터들의 거리를 최소화하거나 다른 군집 간의 거리를 최대화하는 군집 형성  
거리 기반 군집  
계층적 방법, 비계층적 방법 있음  
이상치 탐지, 고객 분류  

## K -Means
비지도 학습 대표적 비계층적 군집 알고리즘  
사전에 정의된 군집 수 K개로 데이터를 분할  
군집의 중심점과 데이터 간의 거리를 계산하여 계층적 군집보다 계산량이 적음.  
K의 최적값을 찾는게 어려움  
Step1. 랜덤하게 K개의 중앙점을 지정
Step2. 유클리드 거리를 이용하여 군집의 중심점과의 거리 계산 => 가장 거리가 가까운 군집으로 할당  
Step3. 군집 내 데이터의 평균을 계산 후 새로운 중심점으로 배정 => 이동 안할 때까지 반복  

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
x, y = make_blobs(random_state=10)
plt.scatter(x[:,0],x[:,1])
```

```python
p1 = [5.683397508810986, 5.055340128019022]
p2 = [6.68000164584778, -5.1538014055317571]
p3 = [-1.3177517986543703, -4.979922671990228]
```

```python
plt.scatter(x[:,0], x[:, 1], s=40, label='Sample')
plt.scatter(p1[0], p1[1], s=100, label='Center 1', c='orange', marker='^')
plt.scatter(p2[0], p2[1], s=100, label='Center 2', c='green', marker='^')
plt.scatter(p3[0], p3[1], s=100, label='Center 3', c='red', marker='^')
plt.legend()
plt.savefig('c1')
```

```python
# 각점에서 중심점까지의 직선거리 계산
import numpy as np
import pandas as pd
dist = pd.DataFrame(x)
dist['c1'] = np.sqrt(np.power(x[:, 0] - p1[0], 2) + np.power(x[:, 1] - p1[1], 2))  # 유클리드 거리 공식
dist['c2'] = np.sqrt(np.power(x[:, 0] - p2[0], 2) + np.power(x[:, 1] - p2[1], 2))
dist['c3'] = np.sqrt(np.power(x[:, 0] - p3[0], 2) + np.power(x[:, 1] - p3[1], 2))
dist
```

```python
# 가장 가까운 중심점 찾기
dist['round1'] = dist.loc[:, 'c1':'c3'].values.argmin(axis=1)
dist
```

```python
# 가장 가까운 중심점과 같은 색으로 칠하기
color = ['orange', 'green', 'red']
for cls in range(3):
    subset = dist.loc[dist['round1']==cls, [0, 1]] # ==> [0,1]이 점의 좌표
    plt.scatter(subset.iloc[:, 0], subset.iloc[:, 1], s=40, c=color[cls])
    print(subset.mean())
plt.scatter(p1[0], p1[1], s=100, label='Center 1', c='orange', marker='^')
plt.scatter(p2[0], p2[1], s=100, label='Center 2', c='green', marker='^')
plt.scatter(p3[0], p3[1], s=100, label='Center 3', c='red', marker='^')
plt.legend()
plt.savefig('c2')
```

```python
# 새로운 중심 좌표 설정
p1 = [2.840507, 4.858404]
p2 = [5.618347, -9.542885]
p3 = [0.089212,  -5.464417]
```

```python
plt.scatter(x[:,0], x[:, 1], s=40, label='Sample')
plt.scatter(p1[0], p1[1], s=100, label='Center 1', c='orange', marker='^')
plt.scatter(p2[0], p2[1], s=100, label='Center 2', c='green', marker='^')
plt.scatter(p3[0], p3[1], s=100, label='Center 3', c='red', marker='^')
plt.legend()
plt.savefig('c1')
```

```python
from sklearn.cluster import KMeans
color = ['orange', 'green', 'red','cyan','black']

K= KMeans(n_clusters=3).fit(x)
plt.scatter(x[:,[0]],x[:,[1]],s=40,label='sample')

centers = K.cluster_centers_

for idx, center in enumerate(centers):
    plt.scatter(center[0],center[1],s=100,label = f'Center {idx+1}', c=color[idx], marker='^')

plt.legend()
plt.show()
```

```python
from sklearn.datasets import make_moons
x,y = make_moons(200,noise=.05,random_state=0)
plt.scatter(x[:,[0]],x[:,[1]],c=y,s=50)
plt.show()
```

```python
labels = KMeans(n_clusters=2, random_state=0).fit_predict(x)
plt.scatter(x[:,0],x[:,1],c=labels, s=50)
plt.show()
```

```python
# DBSCAN 전부 하나로 나옴
from sklearn.cluster import DBSCAN
labels = DBSCAN().fit_predict(x)
plt.scatter(x[:,0],x[:,1],c=labels, s=50)
plt.show()
```

```python
# 정규화
from sklearn.preprocessing import StandardScaler
x_scaled = StandardScaler().fit_transform(x)
x_scaled.mean(axis=0),x_scaled.std(axis=0)
```

```python
labels = DBSCAN().fit_predict(x_scaled)
plt.scatter(x_scaled[:,0],x_scaled[:,1],c=labels, s=50)
plt.show()
```

군집 알고릭즘 평가  
ARI 군집간 거리  
NMI 집단간 거리  
1이 최고, 음수일수록 안좋음

```python
# 군집평가

from sklearn.metrics import accuracy_score
from sklearn.metrics.cluster import adjusted_rand_score
from sklearn.metrics.cluster import normalized_mutual_info_score

x, y = make_blobs(random_state=10)
labels = KMeans(n_clusters=3).fit_predict(x)

print('ACC: ', accuracy_score(y, labels))
print('ARI :', adjusted_rand_score(y, labels))
print('NMI :', normalized_mutual_info_score(y, labels))
```

```python
from sklearn.datasets import load_iris
iris = load_iris()

k_range = range(2,15)
inertias = []

for k in k_range:
    model = KMeans(n_clusters=k).fit(iris['data'])
    inertias.append(model.inertia_)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(k_range, inertias)
plt.xticks(k_range)
plt.grid()
```

```python
# 실루엣 계수 계산
from sklearn.metrics import silhouette_score

k_range = range(2, 15)
scores = []

for k in k_range:
    model = KMeans(n_clusters=k).fit(iris['data'])
    scores.append(silhouette_score(iris['data'], model.labels_))
```

```python
plt.figure(figsize=(10, 5))
plt.plot(k_range, scores)
plt.xticks(k_range)
plt.grid()
```
